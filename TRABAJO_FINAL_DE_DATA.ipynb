```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Cargar tus datos
df = pd.read_csv('supermarket_sales - Sheet1.csv')

# Eliminar columnas innecesarias o no relevantes
df = df.drop(['Invoice ID', 'Date', 'Time'], axis=1)

# Convertir variables categÃ³ricas en variables dummy
df = pd.get_dummies(df, columns=['Gender', 'Product line', 'City', 'Customer type', 'Payment'], drop_first=True)

# Definir las variables predictoras (X) y la variable objetivo (y)
X = df.drop('Branch', axis=1)
y = df['Branch']

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)

# Entrenar el modelo Random Forest
rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)

# Predecir en el conjunto de prueba
y_pred_rfc = rfc.predict(X_test)

# Calcular la precisiÃ³n
accuracy_rfc = rfc.score(X_test, y_test)

# Imprimir matriz de confusiÃ³n
disp_rfc = ConfusionMatrixDisplay(confusion_matrix(y_test, y_pred_rfc), display_labels=df['Branch'].unique())
disp_rfc.plot(cmap='Blues', values_format='d')

# Agregar el porcentaje de precisiÃ³n en el grÃ¡fico
plt.text(0.5, 0.5, f'Accuracy: {round(accuracy_rfc * 100, 2)}%', color='red', fontsize=12, ha='center', va='center', transform=plt.gca().transAxes)

plt.title('Matriz de ConfusiÃ³n para Random Forest')
plt.show()

# Imprimir reporte de clasificaciÃ³n
print("\n--------------------------------------------------------")
print(classification_report(y_test, y_pred_rfc))
print("\n--------------------------------------------------------")
```

MODELO REGRESION LOGISTICA


```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Cargar tus datos
df = pd.read_csv('supermarket_sales - Sheet1.csv')

# Eliminar columnas innecesarias o no relevantes
df = df.drop(['Invoice ID', 'Date', 'Time'], axis=1)

# Convertir variables categÃ³ricas en variables dummy
df = pd.get_dummies(df, columns=['Gender', 'Product line', 'City', 'Customer type', 'Payment'], drop_first=True)

# Definir las variables predictoras (X) y la variable objetivo (y)
X = df.drop('Branch', axis=1)
y = df['Branch']

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)

# Entrenar el modelo de regresiÃ³n logÃ­stica
logreg = LogisticRegression()
logreg.fit(X_train, y_train)

# Predecir en el conjunto de prueba
y_pred_logreg = logreg.predict(X_test)

# Calcular la precisiÃ³n
accuracy_logreg = logreg.score(X_test, y_test)

# Imprimir matriz de confusiÃ³n
disp_logreg = ConfusionMatrixDisplay(confusion_matrix(y_test, y_pred_logreg), display_labels=df['Branch'].unique())
disp_logreg.plot(cmap='Blues', values_format='d')

# Agregar el porcentaje de precisiÃ³n en el grÃ¡fico
plt.text(0.5, 0.5, f'Accuracy: {round(accuracy_logreg * 100, 2)}%', color='red', fontsize=12, ha='center', va='center', transform=plt.gca().transAxes)

plt.title('Matriz de ConfusiÃ³n para RegresiÃ³n LogÃ­stica')
plt.show()

# Imprimir reporte de clasificaciÃ³n
print("\n--------------------------------------------------------")
print(classification_report(y_test, y_pred_logreg))
print("\n--------------------------------------------------------")
```

MODELO SUPPORT VECTOR MACHINE

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay

# Cargar el archivo CSV
file_path = "supermarket_sales - Sheet1.csv"
df = pd.read_csv(file_path)

# Seleccionar las variables predictoras (X) y la variable objetivo (y)
X = df.drop(['Invoice ID', 'Branch', 'City', 'Date', 'Time', 'Payment'], axis=1)  # Reemplaza con el nombre real de tus columnas
y = df['Branch']  # Reemplaza 'Branch' con el nombre real de tu columna de variable objetivo

# Aplicar codificaciÃ³n one-hot a las variables categÃ³ricas
categorical_features = ['Customer type', 'Gender', 'Product line']
X_encoded = pd.get_dummies(X, columns=categorical_features, drop_first=True)

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=2)

# Entrenar el modelo Naive Bayes con GaussianNB
nb = GaussianNB()
nb.fit(X_train, y_train)

# Predecir en el conjunto de prueba
y_pred_nb = nb.predict(X_test)

# Imprimir matriz de confusiÃ³n
disp_nb = ConfusionMatrixDisplay(confusion_matrix(y_test, y_pred_nb), display_labels=df['Branch'].unique())
disp_nb.plot(cmap='Blues', values_format='d')

# Imprimir reporte de clasificaciÃ³n
print("\n--------------------------------------------------------")
print(classification_report(y_test, y_pred_nb))
print("\n--------------------------------------------------------")
```

MODELO DE SUPPORT VECTOR MACHINE


```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Cargar el archivo CSV
file_path = "supermarket_sales - Sheet1.csv"
df = pd.read_csv(file_path)

# Seleccionar las variables predictoras (X) y la variable objetivo (y)
X = df.drop(['Invoice ID', 'Branch', 'City', 'Date', 'Time', 'Payment'], axis=1)
y = df['Branch']

# Aplicar codificaciÃ³n one-hot a las variables categÃ³ricas
categorical_features = ['Customer type', 'Gender', 'Product line']
X_encoded = pd.get_dummies(X, columns=categorical_features, drop_first=True)

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=2)

# Entrenar Support Vector Machine y mostrar matriz de confusiÃ³n
svc = SVC()
svc.fit(X_train, y_train)

# Predecir en el conjunto de prueba
y_pred_svc = svc.predict(X_test)

# Calcular la precisiÃ³n
accuracy_svc = svc.score(X_test, y_test)

# Imprimir matriz de confusiÃ³n
disp_svc = ConfusionMatrixDisplay(confusion_matrix(y_test, y_pred_svc), display_labels=df['Branch'].unique())
disp_svc.plot(cmap='Blues', values_format='d')

# Agregar el porcentaje de precisiÃ³n en el grÃ¡fico
plt.text(0.5, 0.5, f'Accuracy: {round(accuracy_svc * 100, 2)}%', color='red', fontsize=12, ha='center', va='center', transform=plt.gca().transAxes)

plt.title('Matriz de ConfusiÃ³n para Support Vector Machine')
plt.show()

# Imprimir reporte de clasificaciÃ³n
print("\n--------------------------------------------------------")
print(classification_report(y_test, y_pred_svc))
print("\n--------------------------------------------------------")
```
